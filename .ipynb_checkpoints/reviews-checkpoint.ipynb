# Label Encoding

As it is very difficult to teach a model the difference between an extremely positive comment and a positive comment, I decided to encode labels based on three different classes: positive, neutral, negative

```python
#Creating a function to encode labels: 2=positive, 1=neutral, 0=negative
def sentiment(label):
    if label == 5 or label == 4:
        return 2
    if label==3:
        return 1
    else:
        return 0
```

```python
labels=['negative','neutral','positive']
#Creating the new column with the encoded labels
df['label']=df.Rating.apply(lambda x: sentiment(x))
```

# Text Preprocessing

```python
ps = PorterStemmer()
reviews = []
for i in range(0, len(df.Review)):
    if i%1000==0:
        print(i)
    review = re.sub('[^a-zA-Z]', ' ', df.Review[i]) #removing everything which are not letters
    review = review.lower() #lowering strings
    review = review.split() #splitting so that we can iterate
    
    review = [ps.stem(word) for word in review] # stemming applied
    review = ' '.join(review) #joining to recreate the sentence
    reviews.append(review)
```

```python
print(f'Example of review before stemming: {df.Review[0]}')
print('\n')
print(f'Example of review after stemming: {reviews[0]}')
```

# Train/test split

```python
#splitting the daatset into training and validation set.
train_texts, val_texts, train_labels, val_labels = train_test_split(reviews, df.label, test_size=0.2)
```

```python
train_labels.value_counts()
```

# Tokenizing and padding 

```python
token = Tokenizer() #Initialize the tokenizer
token.fit_on_texts(train_texts) #Fit the tokenizer to the reviews

train_text= token.texts_to_sequences(train_texts) #Convert the reviews into sequences for keras to use
test_text= token.texts_to_sequences(val_texts)


max_length = max([len(x) for x in train_text])
vocab_size = len(token.word_index)+1

X_train= pad_sequences(train_text, padding='pre',maxlen=max_length) #Pad the sequences to make them similar lengths
X_val= pad_sequences(test_text, padding='pre',maxlen=max_length)
```

# Model Building

```python
embedding_vector_features=16
model=Sequential()
model.add(Embedding(vocab_size,embedding_vector_features,input_length=max_length))
model.add(Dropout(0.3))
model.add(LSTM(40))
model.add(Dropout(0.3))
model.add(Dense(3,activation='softmax'))
model.compile(loss=SparseCategoricalCrossentropy(),
              optimizer='adam',metrics=['accuracy']
             )
print(model.summary())
```

# Model training

```python
y_train,y_val=train_labels, val_labels
```

```python
model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=3,batch_size=64)
```

# Performance Metrics And Accuracy

```python
y_pred=model.predict_classes(X_val)
```

```python
print(classification_report(y_val, y_pred, target_names = labels))
```

```python
sns.heatmap(confusion_matrix(y_val,y_pred),annot=True,fmt='d',cmap="YlGnBu")
```
